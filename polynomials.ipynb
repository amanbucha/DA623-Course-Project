# Polynomials, Polynomial Bases, and Demonstration

### Polynomials

Polynomials are mathematical expressions consisting of variables and coefficients, combined using addition, subtraction, multiplication, and non-negative integer exponents. They are typically written in the form:

$ P(x) = a_nx^n + a_{n-1}x^{n-1} + \ldots + a_1x + a_0 $

Where:
- $P(x)$  represents the polynomial function.
- $a_n, a_{n-1}, \ldots, a_1, a_0$ are constants called coefficients.
- $x$ is the variable.
- $n$ is a non-negative integer representing the degree of the polynomial.


### Polynomial Bases

Polynomial bases are sets of polynomials that form a basis for a vector space of polynomials. In simpler terms, they are a collection of polynomials that can be used to represent any polynomial of a given degree or less. 


##### Vandermonde Polynomial Basis

The Vandermonde polynomial basis is a specific type of polynomial basis named after the French mathematician Alexandre-Th√©ophile Vandermonde. It consists of polynomials of the form:

$$v_k(x) = x^k$$

where $k$ ranges from 0 to $n$, and $n$ is the degree of the polynomial.

The Vandermonde polynomial basis is particularly useful in interpolation problems. Given a set of $n+1$ distinct points $(x_0, y_0), (x_1, y_1), ..., (x_n, y_n)$, the Vandermonde polynomial basis can be used to construct a unique polynomial of degree $n$ that passes through all of these points.

The polynomial is constructed as a linear combination of the Vandermonde basis polynomials:

$$p(x) = a_0 v_0(x) + a_1 v_1(x) + ... + a_n v_n(x)$$

where $a_0, a_1, ..., a_n$ are coefficients to be determined.

The Vandermonde polynomial basis is particularly advantageous because it simplifies the process of finding the coefficients of the interpolating polynomial. Specifically, the coefficients can be found by solving a system of linear equations, known as the Vandermonde system, which arises naturally from the interpolation problem.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range
x_values = np.linspace(-2, 2, 400)

# Define the degrees of Vandermonde polynomials
degrees = [0, 1, 2, 3, 4]

# Plot the Vandermonde polynomials
plt.figure(figsize=(10, 6))
for degree in degrees:
    vandermonde_poly = x_values ** degree
    plt.plot(x_values, vandermonde_poly, label=f'$x^{degree}$')

plt.title('Vandermonde Polynomials')
plt.xlabel('$x$')
plt.ylabel('$v_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

##### Vandermonde Matrix

The Vandermonde matrix is a square matrix derived from a given set of $n$ data points. It is constructed by forming powers of each data point up to a specified degree. 

Mathematically, given $n$ distinct points $x_1, x_2, \ldots, x_n$ and a degree $d$, the Vandermonde matrix $V$ is defined as:

$$
V = \begin{bmatrix}
1 & x_1 & x_1^2 & \cdots & x_1^d \\
1 & x_2 & x_2^2 & \cdots & x_2^d \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_n & x_n^2 & \cdots & x_n^d
\end{bmatrix}
$$

The Vandermonde matrix is commonly used in polynomial interpolation and curve fitting. Here are some key uses:

1. **Polynomial Interpolation**: Given a set of data points, the Vandermonde matrix can be used to interpolate a polynomial function that passes through these points. By solving the system of linear equations $V \mathbf{c} = \mathbf{y}$, where $\mathbf{c}$ is the coefficient vector of the polynomial and $\mathbf{y}$ is the vector of data points, we can find the coefficients of the polynomial.

2. **Curve Fitting**: The Vandermonde matrix can also be used in curve fitting applications, where we aim to find a polynomial function that best fits a given set of data points. This is typically done by minimizing the sum of squared errors between the observed data points and the corresponding points on the polynomial curve.

3. **Signal Processing**: In signal processing, the Vandermonde matrix is used in methods such as spectral estimation and parameter estimation. It helps in modeling signals with polynomial functions, which is useful in various signal processing tasks including filtering and feature extraction.

4. **Numerical Analysis**: The Vandermonde matrix is used in numerical analysis for polynomial approximation and numerical integration. It plays a crucial role in techniques such as least squares approximation and Gaussian quadrature.

Overall, the Vandermonde matrix is a versatile tool in mathematics and engineering, particularly in tasks involving polynomial functions and approximation. Its simple structure and easy computation make it widely applicable in various domains.

### Orthogonal Polynomial Bases

Orthogonal polynomial bases are sets of polynomials that possess a special property: they are mutually orthogonal with respect to some inner product. 

In simpler terms, if you take any two different polynomials from the basis and compute their inner product (a measure of how "similar" they are), you get zero. Mathematically, for polynomials $P_i(x)$ and $P_j(x)$ from the basis where $i \neq j$, the inner product is zero:

$$\int_{a}^{b} P_i(x) \cdot P_j(x) \cdot w(x)\, dx = 0$$

where $a$ and $b$ are the limits of integration, and $w(x)$ is a weight function corresponding to the orthogonal polynomial bases.

The most common examples of orthogonal polynomial bases are the Legendre polynomials, the Chebyshev polynomials, and the Hermite polynomials.

Orthogonal polynomial bases have several important properties and applications:

1. **Orthogonality**: As mentioned, the polynomials in the basis are orthogonal to each other. This property simplifies various calculations and expansions involving these polynomials.

2. **Weighted Approximation**: Orthogonal polynomial bases can be used for weighted least squares approximation. By choosing an appropriate weight function, one can obtain an optimal approximation for a given set of data.

3. **Numerical Integration**: Orthogonal polynomials are often used in numerical integration methods, such as Gaussian quadrature. The zeros of orthogonal polynomials can be used as the nodes for these quadrature formulas, leading to highly accurate integration schemes.

4. **Signal Processing**: In signal processing, orthogonal polynomials find applications in areas like spectral analysis, filtering, and compression.

Orthogonal polynomial bases are an essential tool in many areas of mathematics and its applications, providing efficient ways to handle polynomial approximations, integration problems, and signal analysis.


##### Legendre Polynomials

Legendre polynomials are a set of orthogonal polynomials defined on the interval $[-1, 1]$. They are named after the French mathematician Adrien-Marie Legendre, who first introduced them in 1782. Legendre polynomials play a significant role in various fields of science and engineering, particularly in solving problems involving spherical harmonics, boundary value problems in partial differential equations, and in numerical methods like spectral methods.

The Legendre polynomials, denoted as $P_n(x)$, are defined recursively using Rodrigues' formula:

$$P_n(x) = \frac{1}{2^n n!} \frac{d^n}{dx^n} (x^2 - 1)^n$$

where $n$ is a non-negative integer, and $\frac{d^n}{dx^n}$ denotes the nth derivative.

The first few Legendre polynomials are:

- $P_0(x) = 1$
- $P_1(x) = x$
- $P_2(x) = \frac{1}{2}(3x^2 - 1)$
- $P_3(x) = \frac{1}{2}(5x^3 - 3x)$
- $P_4(x) = \frac{1}{8}(35x^4 - 30x^2 + 3)$
- ...

Legendre polynomials have several key properties:

1. **Orthogonality**: Legendre polynomials are orthogonal with respect to the weight function $w(x) = 1$, meaning that their inner product over the interval $[-1, 1]$ is zero for distinct polynomials:

   $$\int_{-1}^{1} P_m(x) P_n(x) \, dx = 0$$

   when $m \neq n$.

2. **Normalization**: The Legendre polynomials are normalized such that $P_n(1) = 1$ for all $n$.

3. **Recurrence Relation**: Legendre polynomials can be generated using a recurrence relation:

   $$(n+1)P_{n+1}(x) = (2n+1)xP_n(x) - nP_{n-1}(x)$$

   where $P_{-1}(x) = 0$ and $P_0(x) = 1$.

4. **Zeros**: The zeros of Legendre polynomials are important, as they are used as nodes in numerical integration methods like Gaussian quadrature.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

# Define x values
x = np.linspace(-1, 1, 400)

# Define Legendre polynomials up to degree 4
legendre_0 = legendre(0)(x)
legendre_1 = legendre(1)(x)
legendre_2 = legendre(2)(x)
legendre_3 = legendre(3)(x)
legendre_4 = legendre(4)(x)

# Plot Legendre polynomials
plt.figure(figsize=(10, 6))
plt.plot(x, legendre_0, label='$P_0(x)$')
plt.plot(x, legendre_1, label='$P_1(x)$')
plt.plot(x, legendre_2, label='$P_2(x)$')
plt.plot(x, legendre_3, label='$P_3(x)$')
plt.plot(x, legendre_4, label='$P_4(x)$')
plt.title('Legendre Polynomials')
plt.xlabel('$x$')
plt.ylabel('$P_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

Suppose we have the polynomial:

$$p(x) = 2x^3 - 3x^2 + 4x - 1$$

We want to express this polynomial as a linear combination of Legendre polynomials. To do this, we'll need to compute the coefficients of each Legendre polynomial term.

The Legendre polynomials are orthogonal over the interval $[-1, 1]$, so we can use the following formula to compute the coefficients:

$$a_n = \frac{2n + 1}{2} \int_{-1}^{1} p(x) \cdot P_n(x) \, dx$$

where $P_n(x)$ denotes the nth Legendre polynomial.

Let's break down $p(x)$ into Legendre polynomials up to degree 3:

1. $P_0(x) = 1$
2. $P_1(x) = x$
3. $P_2(x) = \frac{1}{2}(3x^2 - 1)$
4. $P_3(x) = \frac{1}{2}(5x^3 - 3x)$

Now, let's compute the coefficients $a_0$, $a_1$, $a_2$, and $a_3$ using the above formula:

$$
\begin{align*}
a_0 & = \frac{1}{2} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot 1 \, dx \\
a_1 & = \frac{3}{2} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot x \, dx \\
a_2 & = \frac{5}{2} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot \frac{1}{2}(3x^2 - 1) \, dx \\
a_3 & = \frac{7}{2} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot \frac{1}{2}(5x^3 - 3x) \, dx \\
\end{align*}
$$

We can evaluate these integrals numerically to find the coefficients $a_0$, $a_1$, $a_2$, and $a_3$. Once we have these coefficients, we can express $p(x)$ as a linear combination of Legendre polynomials:

$$p(x) = a_0 P_0(x) + a_1 P_1(x) + a_2 P_2(x) + a_3 P_3(x)$$

This represents $p(x)$ in terms of Legendre polynomials up to degree 3. We can similarly extend this process to higher degrees if needed.

Below is a python script which can express a polynomial as a linear combination of Legendre polynomials:


```python
import numpy as np
from scipy.integrate import quad
from scipy.special import legendre

def polynomial_to_legendre(polynomial, degree):
    coefficients = []
    legendre_polynomials = [legendre(n) for n in range(degree + 1)]
    
    for n in range(degree + 1):
        integrand = lambda x: polynomial(x) * legendre_polynomials[n](x)
        coefficient, _ = quad(integrand, -1, 1)
        coefficient *= (2 * n + 1) / 2
        coefficients.append(coefficient)
    
    legendre_polynomials = [legendre(n) for n in range(degree + 1)]
    legendre_terms = [f'{coefficients[n]:.4f} * P_{n}(x)' for n in range(degree + 1)]
    legendre_expression = ' + '.join(legendre_terms)
    print(f'\np(x) = {legendre_expression}')


# Example polynomial function: 2x^3 - 3x^2 + 4x - 1
def polynomial(x):
    return 2 * x**3 - 3 * x**2 + 4 * x - 1
degree = 3

coefficients = polynomial_to_legendre(polynomial, degree)
```

##### Chebyshev polynomials

Chebyshev polynomials are a family of orthogonal polynomials that arise in various areas of mathematics and engineering. They are named after the Russian mathematician Pafnuty Chebyshev. These polynomials have numerous important properties and applications, particularly in numerical analysis, approximation theory, signal processing, and physics.

The Chebyshev polynomials of the first kind, denoted as $T_n(x)$, are defined recursively by the following formula:

$$T_0(x) = 1$$
$$T_1(x) = x$$
$$T_{n+1}(x) = 2xT_n(x) - T_{n-1}(x)$$

for $n \geq 1$.

The first few Chebyshev polynomials are:

- $T_0(x) = 1$
- $T_1(x) = x$
- $T_2(x) = 2x^2 - 1$
- $T_3(x) = 4x^3 - 3x$
- $T_4(x) = 8x^4 - 8x^2 + 1$
- ...

Chebyshev polynomials have several key properties:

1. **Orthogonality**: The Chebyshev polynomials are orthogonal with respect to the weight function $w(x) = \frac{1}{\sqrt{1 - x^2}}$ on the interval $[-1, 1]$, meaning that their inner product is zero for distinct polynomials:

   $$\int_{-1}^{1} T_m(x) \cdot T_n(x) \frac{1}{\sqrt{1 - x^2}} \, dx = 0$$

   when $m \neq n$.

2. **Recurrence Relation**: As shown in the definition, Chebyshev polynomials can be generated recursively using a recurrence relation, making it efficient to compute higher-degree polynomials.

3. **Roots**: The roots of Chebyshev polynomials have interesting properties. For $T_n(x)$, the $n$ roots are given by:

   $$x_k = \cos\left(\frac{(2k - 1)\pi}{2n}\right)$$

   for $k = 1, 2, ..., n$.

4. **Minimax Property**: Chebyshev polynomials are optimal in the sense that they minimize the maximum absolute deviation from zero among all monic polynomials of the same degree on the interval $[-1, 1]$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import chebyt

# Define x values
x = np.linspace(-1, 1, 400)

# Define Chebyshev polynomials up to degree 4
chebyshev_0 = chebyt(0)(x)
chebyshev_1 = chebyt(1)(x)
chebyshev_2 = chebyt(2)(x)
chebyshev_3 = chebyt(3)(x)
chebyshev_4 = chebyt(4)(x)

# Plot Chebyshev polynomials
plt.figure(figsize=(10, 6))
plt.plot(x, chebyshev_0, label='$T_0(x)$')
plt.plot(x, chebyshev_1, label='$T_1(x)$')
plt.plot(x, chebyshev_2, label='$T_2(x)$')
plt.plot(x, chebyshev_3, label='$T_3(x)$')
plt.plot(x, chebyshev_4, label='$T_4(x)$')
plt.title('Chebyshev Polynomials')
plt.xlabel('$x$')
plt.ylabel('$T_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

Let us focus on the polynomial again:

$$p(x) = 2x^3 - 3x^2 + 4x - 1$$

We want to express this polynomial as a linear combination of Chebyshev polynomials. To do this, we'll need to compute the coefficients of each Chebyshev polynomial term.

The Chebyshev polynomials of the first kind, denoted as $T_n(x)$, are defined recursively by the following formula:

$$T_0(x) = 1$$
$$T_1(x) = x$$
$$T_{n+1}(x) = 2xT_n(x) - T_{n-1}(x)$$

for $n \geq 1$.

Now, let's compute the coefficients $a_0$, $a_1$, $a_2$, and $a_3$ using the following formula:

$$a_n = \frac{2}{\pi} \int_{-1}^{1} p(x) \cdot T_n(x) \frac{1}{\sqrt{1 - x^2}} \, dx$$

where $T_n(x)$ denotes the nth Chebyshev polynomial.

Let's break down $p(x)$ into Chebyshev polynomials up to degree 3:

1. $T_0(x) = 1$
2. $T_1(x) = x$
3. $T_2(x) = 2x^2 - 1$
4. $T_3(x) = 4x^3 - 3x$

Now, let's compute the coefficients $a_0$, $a_1$, $a_2$, and $a_3$ using the above formula:

$$
\begin{align*}
a_0 & = \frac{2}{\pi} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot 1 \frac{1}{\sqrt{1 - x^2}} \, dx \\
a_1 & = \frac{2}{\pi} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot x \frac{1}{\sqrt{1 - x^2}} \, dx \\
a_2 & = \frac{2}{\pi} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot (2x^2 - 1) \frac{1}{\sqrt{1 - x^2}} \, dx \\
a_3 & = \frac{2}{\pi} \int_{-1}^{1} (2x^3 - 3x^2 + 4x - 1) \cdot (4x^3 - 3x) \frac{1}{\sqrt{1 - x^2}} \, dx \\
\end{align*}
$$

We can evaluate these integrals numerically to find the coefficients $a_0$, $a_1$, $a_2$, and $a_3$. Once we have these coefficients, we can express $p(x)$ as a linear combination of Chebyshev polynomials:

$$p(x) = a_0 T_0(x) + a_1 T_1(x) + a_2 T_2(x) + a_3 T_3(x)$$

This represents $p(x)$ in terms of Chebyshev polynomials up to degree 3. We can similarly extend this process to higher degrees if needed.

Below is a python script which can express a polynomial as a linear combination of Chebyshev polynomials:


```python
import numpy as np
from scipy.integrate import quad
from scipy.special import chebyt

def polynomial_to_chebyshev(polynomial, degree):
    coefficients = []
    chebyshev_polynomials = [chebyt(n) for n in range(degree + 1)]
    
    for n in range(degree + 1):
        integrand = lambda x: polynomial(x) * chebyshev_polynomials[n](x)
        coefficient, _ = quad(integrand, -1, 1)
        coefficient *= (2 * n + 1) / 2
        coefficients.append(coefficient)
    
    chebyshev_terms = [f'{coefficients[n]:.4f} * P_{n}(x)' for n in range(degree + 1)]
    chebyshev_expression = ' + '.join(chebyshev_terms)
    print(f'\np(x) = {chebyshev_expression}')


# Example polynomial function: 2x^3 - 3x^2 + 4x - 1
def polynomial(x):
    return 2 * x**3 - 3 * x**2 + 4 * x - 1
degree = 3

coefficients = polynomial_to_chebyshev(polynomial, degree)
```
